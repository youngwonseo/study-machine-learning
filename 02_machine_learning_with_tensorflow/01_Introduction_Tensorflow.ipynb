## 1.1 텐서플로우란?

텐서플로우는 딥러닝 프레임워크로써 이름에서와 같이 텐서(Tensor)를 흘려(flow)보내면서 데이터를 처리합니다. 여기서 텐서는 다차원(n-dimensonal) 행열을 의미하며 흘려보내는것은 텐서에 대한 연산을 의미합니다.

텐서플로우는 이것을 위해 계산그래프로 모델링합니다.

실습환경을 위해 구글 드라이브에서 제공한는 ipython환경인 Colaboratory를 사용합니다.

## 1.2 계산그래프(computional graph)
텐서를 흘려보낸다고 했는데 어떤식으로 흘려보낼까요? 텐서플로우는 계산그래프를 구성해서 텐서를 흘려보냅니다.다음 그림을 살펴봅시다.

![](https://youngwonseo.github.io/public/deeplearning/computational_graph1.png)


인풋으로 상수 11과 3이 입력되고 add연산을 수행한 아웃풋을 출력하는 계산그래프입니다. 텐서플로우 코드로 표현해 봅시다.

```
import tensorflow as tf

input1 = tf.constant(11)
input2 = tf.constant(3)

output = tf.add(input1, input2)
```

위 코드는 명확합니다. 첫 번째 tensorflow를 임포트 하고 상수 11과 3, 그리고 add연산을 수행해 output을 출력합니다. 한번 출력해 보겠습니다.

```
print(output)
```

보기와 다르게 출력이 되지 않습니다. 이유는 위 코드는 계산그래프를 생성하는 그래프의 실행은 session을 통해 이루어집니다.




```
with tf.Session() as sess:
  print("11 + 3 = %d "% sess.run(output))
```

이처럼 텐서플로우는 다음의 두 과정을 거처 계산그래프를 실행합니다.
* 그래프 생성(정의)
* 그래프 실행

또한 계산그래프의 그림에서 알수 있듯이 엣지와 노드는 다음을 나타냅니다.

* edge는 multidimensional arrays, 즉 텐서(Tensors)
* node는 operation


첫 번째 예제에서는 2개의 입값이 모두 상수(tf.constant)였습니다. 그럼 입력데이터나 변수는 어떻게 설정하는지 알아보겠습니다.

### Placeholder

텐서플로우는 계산그래프를 정의할때 입력을 edge로 구성해야 합니다. 그렇다고 여러가지 데이터를 상수로 정의하여 사용하지는 않습니다. 대신 placeholder를 사용해서 입력할 데이터가 담길수 있는 텐서플로우에서 제공하는 자료구조를 사용하여 계산그래프를 구성합니다.

다음과 같은 계산그래프를 정의하고 실행해 봅시다.

![](https://youngwonseo.github.io/public/deeplearning/computational_graph2.png)


```
import tensorflow as tf

x1 = tf.placeholder("float")
x2 = tf.placeholder("float")

const3 = tf.constant(3.)

mul_result = tf.multiply(x1,x2)
output = tf.add(mul_result, const3)
```

```
with tf.Session() as sess:
  print("result : %f " % sess.run(output, feed_dict={x1: 5, x2: 2}) )
```

### 계산 그래프의 장점
* 여러개의 분산된 디바이스에서 병렬(simultaneously) 실행가능
* 여러단계의 연산을 순차적으로 정의하고 변경이 쉬움
* 딥러닝 프레임워크에서 경사하강법을 자동적으로 적용
